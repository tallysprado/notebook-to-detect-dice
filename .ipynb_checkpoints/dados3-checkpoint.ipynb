```python
#Não foi utilizado clusterização
#A detecção de blobs é utilizada apenas para contar os pontos
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('dados3.png')
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = rgb_img.copy()
plt.imshow(image)
plt.title('dados3.png')
plt.show()
```

```python
#Filtros necessários, a ordem destes diferem de dados1 e dados2
ret, image = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY_INV)
image = cv2.blur(image, (3,3))
kernel = np.ones((5,5), np.uint8)
image = cv2.erode(image, kernel, iterations=1)
plt.imshow(image)
plt.title('dados3.png com filtros')
plt.show()
```

```python
#inicializando blob detector apenas para contagem dos pontos de cada dado único
params = cv2.SimpleBlobDetector_Params()

params.filterByArea = True
params.filterByColor = True
params.minArea = 1
params.minDistBetweenBlobs = 1
params.blobColor = 0
detector = cv2.SimpleBlobDetector_create(params)
```

```python
#Utilizando filltro de Canny e contour detection
#O CMAP DE MATPLOTLIB É DIFERENTE DE RGB
canny = cv2.Canny(image, 100, 550)
out = image.copy()
original = cv2.imread('dados3.png')
cnts, hier = cv2.findContours(
    canny.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE
)
array_dices = []
for c in cnts:
    x, y, w, h = cv2.boundingRect(c)
    if w > 40:
        new_img = out[y:y+h, x:x+w]
        keypoints = detector.detect(new_img)
        img_with_keypoints = cv2.drawKeypoints(
        image, keypoints, np.array([]), (255, 0, 0),
        cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
        )
        array_dices.append(new_img)
        font = cv2.FONT_HERSHEY_SIMPLEX
        bottomLeft = (x,y)
        fontScale = 1
        fontColor = (255,0,0)
        lineType = 2
        cv2.putText(
            original, str(len(keypoints)), bottomLeft, font, fontScale, fontColor, lineType
        )      
        cv2.rectangle(original, (x, y), (x+w, y+h), (255, 0, 0), 1)
plt.imshow(original)
plt.title('dados3.png')
plt.show()
```
